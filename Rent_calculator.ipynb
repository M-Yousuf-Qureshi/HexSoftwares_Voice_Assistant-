# **Rent** **calculator**




First import the libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Import the dataset from Kaggle
data = pd.read_csv("/content/Makaan_data_700pages.csv")
```

```python
data
```

# Generate a list of the last 10 months

```python
import datetime
end_date = datetime.date.today().replace(day=1) # Start from the first day of the current month
months = []
for i in range(10):
    current_month_start = (end_date - datetime.timedelta(days=30*i)).replace(day=1)
    months.append(current_month_start)
months.reverse() # Order from oldest to newest

# Assign a random month to each row in the dataset
data['Month'] = np.random.choice(months, size=len(data))

# Display a sample of the data with the new 'Month' column
display(data[['Property_type', 'Location', 'Cleaned_Rent_Price', 'Month']].head())
```

# Calculate the average rent per month

```python
society_rent_per_month = data.groupby('Month')['Cleaned_Rent_Price'].mean().sort_index()

print("Average Rent per Month (simulated 'Society Rent'):")
display(society_rent_per_month)
```

### Calculate the Rent

```python
def clean_rent_price(price_str):
    if isinstance(price_str, str):
        price_str = price_str.replace(',', '')
        if 'L' in price_str:
            return float(price_str.replace('L', '').strip()) * 100000
        try:
            return float(price_str)
        except ValueError:
            return np.nan # Handle cases where conversion fails
    return price_str

data['Cleaned_Rent_Price'] = data['Rent_price'].apply(clean_rent_price)

# Replace 'ApartmentApartment' with 'Apartment' in 'Property_type' column
data['Property_type'] = data['Property_type'].replace('ApartmentApartment', 'Apartment')

# Group by 'Property_type' and calculate the average 'Cleaned_Rent_Price'
average_rent_per_class = data.groupby('Property_type')['Cleaned_Rent_Price'].mean().sort_values(ascending=False)

print("Average Rent per Property Type:")
display(average_rent_per_class)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.lineplot(x=society_rent_per_month.index, y=society_rent_per_month.values)
plt.title('Average Society Rent per Month (Simulated)')
plt.xlabel('Month')
plt.ylabel('Average Rent')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# Analyze how Location affects the average society rent

```python
# Calculate the average rent per Location
average_rent_by_location = data.groupby('Location')['Cleaned_Rent_Price'].mean().sort_values(ascending=False)

print("Average Rent per Location:")
display(average_rent_by_location.head(10)) # Display top 10 locations by average rent
print("\nAverage Rent per Location (bottom 10):")
display(average_rent_by_location.tail(10)) # Display bottom 10 locations by average rent
```

# Total_average_rent

```python
total_average_rent_sum = average_rent_by_location.sum()

print(f"The sum of the average rents for all locations is: {total_average_rent_sum:,.2f}")
```

# average_rent property type using a bar chart

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Sort locations bytotal_average_rent for top and bottom N
top_n_locations = average_rent_by_location.head(20)
bottom_n_locations = average_rent_by_location.tail(20)

# Create a figure with two subplots
fig, axes = plt.subplots(2, 1, figsize=(12, 16))

# Bar chart for Top N locations
sns.barplot(x=top_n_locations.values, y=top_n_locations.index, ax=axes[0], palette='viridis', hue=top_n_locations.index, legend=False)
axes[0].set_title('Top 20 Locations by Average Rent')
axes[0].set_xlabel('Average Rent')
axes[0].set_ylabel('Location')
axes[0].ticklabel_format(style='plain', axis='x') # Prevent scientific notation

# Bar chart for Bottom N locations
sns.barplot(x=bottom_n_locations.values, y=bottom_n_locations.index, ax=axes[1], palette='plasma', hue=bottom_n_locations.index, legend=False)
axes[1].set_title('Bottom 20 Locations by Average Rent')
axes[1].set_xlabel('Average Rent')
axes[1].set_ylabel('Location')
axes[1].ticklabel_format(style='plain', axis='x') # Prevent scientific notation

plt.tight_layout()
plt.show()
```
